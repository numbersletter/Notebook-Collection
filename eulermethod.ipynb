```python
import numpy as np
from matplotlib import pyplot as plt
from sympy import var
from sympy import sympify
from sympy.utilities.lambdify import lambdify
```

### Euler's Method
Let 
$$ y' = f(x,y)$$
Recall, we can use tangent line approximation to estimate $y(x)$. For example, $$y(x_1) \approx (x_1-x_0)f(x_0,y_0) + y_0 = \Delta x * f(x_0,y_0) + y_0. $$ We can use this $y_1$ to approximate $y_2$: $$y(x_2) = y_2 \approx \Delta x * f(x_1,y_1) + y_1$$ 

```python
x = var('x')
y = var('y')
```

```python
user_in = input("Enter dy/dx: ")
x0 = float(input("Enter x0: "))
y0 = float(input("Enter y0: "))
xf = float(input("Enter final x: "))
n = int(input("Enter number of equally spaced points: "))
```

Make user input function accessible by numpy:

```python
expr = sympify(user_in)
f = lambdify((x,y), expr)
```

Define $\Delta x$

```python
deltax = (xf - x0) / (n - 1)
```

Create two vector of length n which stores all x values:

```python
x_all = np.linspace(x0, xf, n)
```

Define the vector of length n for all y values:

```python
y_all = np.zeros([n])
y_all[0] = y0
```

Use Euler's method in a loop:

```python
for i in range(1, n):
    y_all[i] = deltax * (f(x_all[i-1],y_all[i-1])) + y_all[i-1]
```

Print

```python
for i in range(1, n):
    print(x_all[i], ",", y_all[i])
```

Graph approximation

```python
plt.plot(x_all, y_all, 'o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Euler's Method Approximation of y")
plt.text(xf, y_all[n-1], '({}, {})'.format(xf, round(y_all[n-1], 4)))
plt.show()
```
